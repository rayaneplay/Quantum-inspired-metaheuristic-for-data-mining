## Implementing different distance methods

```python
def distance_manatthan(centroid, datapoint):
    dist = sum(abs(c - d) for c, d in zip(centroid, datapoint))
    return dist

def distance_minkowski(centroid, datapoint, p):
    dist = sum((abs(c - d)**p) for c, d in zip(centroid, datapoint))
    return (dist ** (1/p))


def distance_euclidienne (centroid, datapoint):
    dist = sum(((c - d)**2) for c, d in zip(centroid, datapoint))
    return math.sqrt(dist)
    

def distance_cosine(centroid, datapoint):
    for c,d in zip(centroid, datapoint):
        up = sum (c * d)
        down1 = sum(c * c)
        down2 = sum(d * d)
    dist = 1 - (up / (math.sqrt(down1) * math.sqrt(down2)))
    return dist

def distance_hamming(centroid, datapoint):
    dist =0
    for c,d in zip(centroid, datapoint):
        if c != d:
            dist += 1
    return dist
```

## Creating clusters from random centroids already generated and with chosen distance method

```python

def create_clusters(centroids, dataset,distance_method):
    distances = {}

    for data_point in dataset:
        min_distance = float('inf')  # Initialization to find the minimum distance
        closest_centroid = None

        for centroid in centroids:
            #dist = sum(abs(c - d) for c, d in zip(centroid, data_point))
            dist = distance_method(centroid, data_point)
            if dist < min_distance:
                min_distance = dist
                closest_centroid = tuple(centroid)  # Convert list to tuple to use as a key

        if closest_centroid in distances:
            distances[closest_centroid].append(data_point)  # Add instance and its centroid to the list
        else:
            distances[closest_centroid] = [data_point]  # Create a new list with the instance and its centroid

    return distances


clusters = create_clusters(centroids, data, distance_manatthan)

clusters
```

```python
def update_centroid(clusters):
    new_centroids = []
    for centroid in clusters.keys():
        # for i in range (len(clusters.values[cluster])):
        # for instance in clusters[centroid]:
        new = np.array(clusters[centroid]).T
        means_array = [np.mean(new[0]), np.mean(new[1]),np.mean(new[2]), np.mean(new[3])]
        new_centroids.append(means_array)
    new_centroids = [[float(f'{num:.3f}') for num in sublist] for sublist in new_centroids]
    return new_centroids

new =update_centroid(clusters)

print(new)
```

## Verify if old cluster is the same as new cluster (except for centroids)

```python
def check_for_convergence(new_cluster, old_cluster):
    if old_cluster == {}:
        return False
    else:
        test1 = dic_format(new_cluster)
        test2 = dic_format(old_cluster)
        
        new_keys = set(test1.keys())
        old_keys = set(test2.keys())

        common_keys = new_keys.intersection(old_keys)

        for key in common_keys:
            new_instances = set(map(tuple, test1[key]))
            old_instances = set(map(tuple, test2[key]))

            if new_instances != old_instances:
                return False  # Found a difference, return False

        return True  # All instances in corresponding keys are the same


result = check_for_convergence(test1, test2)
print(result) 
```

## Convert Clusters into an array of class Labels from the original dataset

```python
def find_class(input_vector, dataset):
    # Convert the input vector to a list of strings
    input_str = [str(elem) for elem in input_vector]

    # Iterate through each instance in the dataset
    for instance in dataset:
        # Convert the instance to a list of strings and remove the class label ('\n')
        instance_str = instance[:-1]

        # Check if the string representation of the input vector matches the current instance
        if all(x == y for x, y in zip(input_str, instance_str)):
            # Return the class label of the matched instance
            return instance[-1].strip()  # Assuming the class label is at the end of each instance

    # If no match is found, return None or a message indicating no match
    return "No matching instance found"


def check_classes_cluster(clusters, dataset):
    cluster_classes = []
    for centroid in clusters:
        cluster_class= []
        for point in clusters[centroid]:
            cluster_class.append(find_class(point, dataset))
        cluster_classes.append(cluster_class)
    
    return cluster_classes


```

## Plotting the resulting classes of each cluster into Pie Charts

```python
import matplotlib.pyplot as plt
from collections import Counter



def plot_class_distribution(cluster_classes):
    # Define the classes in the dataset
    classes_in_dataset = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

    num_clusters = len(cluster_classes)

    fig, axs = plt.subplots(1, num_clusters, figsize=(15, 5))

    for i in range(num_clusters):
        class_distribution = Counter(cluster_classes[i])
        total_count = sum(class_distribution.values())

        percentages = [(class_distribution[cls] / total_count) * 100 if cls in class_distribution else 0 for cls in classes_in_dataset]
        non_zero_percentages = [p for p in percentages if p != 0]
        non_zero_classes = [cls for idx, cls in enumerate(classes_in_dataset) if percentages[idx] != 0]

        axs[i].pie(non_zero_percentages, labels=non_zero_classes, autopct='%1.1f%%', startangle=140)
        axs[i].set_title(f'Cluster {i + 1}: Class Distribution')

    plt.tight_layout()
    plt.show()


# Example usage
# Assuming 'cluster_classes' is the array of arrays containing classes in each cluster
plot_class_distribution(cluster_classes)

```
