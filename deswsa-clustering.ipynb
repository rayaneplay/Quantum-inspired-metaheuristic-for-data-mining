```python
from utils import *
from deswsa import *
```

```python
X = load_data("dataset.csv")
X = normalize_data(X)
```

```python
num_elephants = 100
num_clusters_max = 5
p = 0.5
t_max = 100
w_min = 0.5
w_max = 0.8
```

```python
deswsa = DESWSA(X ,num_elephants, p, t_max, w_min, w_max, num_clusters_max)
best_solutions = deswsa.run()
for num_clusters, (best_solution, best_fitness) in best_solutions.items():
    print(f"Meilleure solution trouvée avec {num_clusters} clusters:", best_solution)
    print(f"Meilleure fitness avec {num_clusters} clusters:", best_fitness)
    print("\n")
    centroids = deswsa.get_centroids(best_solution)
    clusters = deswsa.create_clusters(centroids, distance_method=distance_euclidean)
    plot_clusters(clusters, centroids)
```
